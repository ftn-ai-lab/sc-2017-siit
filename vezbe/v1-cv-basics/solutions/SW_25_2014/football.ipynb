```python
import matplotlib.pyplot as plt  
# iscrtavanje slika i plotova unutar samog browsera
%matplotlib inline 

import matplotlib.pylab as pylab
# prikaz vecih slika 
pylab.rcParams['figure.figsize'] = 16,12

import numpy as np
import cv2 # OpenCV biblioteka
```

```python
img_football = cv2.cvtColor(cv2.imread('../../images/football.jpg'), cv2.COLOR_BGR2RGB)
#plt.imshow(img_barcode)
img_football_gs = cv2.cvtColor(img_football, cv2.COLOR_RGB2GRAY) # konvert u grayscale
plt.imshow(img_football_gs, 'gray')
```

```python
ret, img_football_bin = cv2.threshold(img_football_gs, 215, 255, cv2.THRESH_BINARY)
#img_football_bin = cv2.adaptiveThreshold(img_football_gs, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 35, 10)
plt.imshow(img_football_bin, 'gray')
```

```python
img, contours, hierarchy = cv2.findContours(img_football_bin, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)
# img = img_football.copy()
# cv2.drawContours(img, contours, -1, (255, 0, 0), 1)
# plt.imshow(img)

contours_football = [] 
for contour in contours: 
    area = cv2.contourArea(contour)
    if area > 35:
        contours_football.append(contour)

img = img_football.copy()
cv2.drawContours(img, contours_football, -1, (255, 0, 0), 1)
plt.imshow(img)
print 'Ukupan broj belih igraca: %d' % len(contours_football)
```
